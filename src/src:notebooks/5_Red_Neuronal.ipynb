```python
import numpy as np
import pandas as pd
pd.options.display.max_columns = 36
import io 

from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from sklearn.neural_network import MLPClassifier 

from sklearn.metrics import accuracy_score, f1_score, precision_score, roc_auc_score
```

**Datos**

```python
from google.colab import files 
uploaded = files.upload()
```

```python
X_train = pd.read_csv(io.BytesIO(uploaded['X_train.csv']), index_col = 0) 
X_train.head()
```

```python
y_train = pd.read_csv(io.BytesIO(uploaded['y_train.csv']), index_col = 0) 
y_train = y_train.squeeze()
y_train.head()
```

```python
X_test = pd.read_csv(io.BytesIO(uploaded['X_test.csv']), index_col = 0) 
X_test.head()
```

```python
y_test = pd.read_csv(io.BytesIO(uploaded['y_test.csv']), index_col = 0) 
y_test = y_test.squeeze()
y_test.head()
```

**M√©tricas**

```python
def metricas_print(prediccion, y):

    print("Accuracy", accuracy_score(y, prediccion))
    print("Precision", precision_score(y, prediccion))
    print("F1", f1_score(y, prediccion))
    print("AUC", roc_auc_score(y, prediccion))
```

**Red** **Neuronal** - **Multi Layer Perceptron**

```python
# Estandarizamos los datos
scaler = StandardScaler()
scaler.fit(X_train)

X_train_scal = scaler.transform(X_train)
X_test_scal = scaler.transform(X_test)
```

Pruebo relu o tanh en diferentes celdas de c√≥digo porque cuando aumentaba el n√∫mero de par√°metros, tardaba demasiado y nunca llegaba a terminar de ejecutarse

```python
%time
# Multilayer Perceptron Classifier --> Activation: RELU

mlp_params = {"max_iter": [400, 500, 600],
              "activation": ["relu"]}

mlp = GridSearchCV(MLPClassifier(), mlp_params,
                        cv = 10, 
                        scoring= "accuracy",
                        verbose = 10,
                        n_jobs = -1)

mlp.fit(X_train_scal, y_train)

mlp.best_params()
```

```python
mlp.get_params()
```

```python
print("Score Train:", mlp.score(X_train_scal, y_train))
print("Score Test:", mlp.score(X_test_scal, y_test))
```

```python
print("Train scores:")
metricas_print(y_train, mlp.predict(X_train_scal))
```

```python
print("Test scores:")
metricas_print(y_test, mlp.predict(X_test_scal))
```

```python
# Multilayer Perceptron Classifier --> Activation: TANH
# Aunque en train nos da mejor resultado que la de arriba, vemos m√°s diferencia entre train y test(overfitting) --> nos quedamos con funci√≥n activaci√≥n relu)

mlp_params_2 = {"max_iter": [400, 500, 600],
              "activation": ["tanh"]}

mlp_2 = GridSearchCV(MLPClassifier(), mlp_params_2,
                        cv = 10, 
                        scoring= "accuracy",
                        verbose = 10,
                        n_jobs = -1)

mlp_2.fit(X_train_scal, y_train)

print("Par√°metros:", mlp_2.get_params())

print("Score Train:", mlp_2.score(X_train_scal, y_train))
print("Score Test:", mlp_2.score(X_test_scal, y_test))
```

```python
print("Train scores:")
metricas_print(y_train, mlp_2.predict(X_train_scal))
```

```python
print("Test scores:")
metricas_print(y_test, mlp_2.predict(X_test_scal))
```
